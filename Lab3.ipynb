```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = pd.read_csv("accord_sedan.csv")
print(f)
```

```python
f.plot(kind = 'scatter', x = 'price', y = 'mileage')
f.plot(kind = 'density', x = 'price', y = 'mileage')
f.plot(kind = 'hist', x = 'price', y = 'mileage')
f.plot()
```

```python
f.describe()[['price','mileage']]
```

```python
f['isOutlierPrice'] = 1
f['isOutlierMileage'] = 1
```

```python
p_m = f.price.mean()
p_s = f.price.std()
m_m = f.mileage.mean()
m_s = f.mileage.std()
f['isOutlierPrice'] = np.where(abs(f['price'] - p_m) > 2 * p_s, 1, 0)
print(f['isOutlierPrice'])
f['isOutlierMileage'] = np.where(abs(f['mileage'] - m_m) > 2 * m_s, 1, 0)
print(f['isOutlierMileage'])
```

```python
plt.scatter(f['price'], f['mileage'], cmap = 'viridis')
plt.scatter(f['isOutlierPrice'], f['isOutlierMileage'], s = 15, c = 'grey')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats

f = pd.read_csv("TB_burden_countries_2014-09-29.csv", encoding = 'latin-1')
df = pd.DataFrame(f)
df2 = df.fillna(0)
```

```python
h = sorted(df2['e_inc_num_hi'])
res = stats.norm.pdf(h, np.mean(h), np.std(h))
fig = sm.qqplot(res, stats.t, fit = True, line = '45')
plt.show()
```

```python
import numpy as np
from scipy.stats import kurtosis
from scipy.stats import skew
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

ran_1 = np.random.normal(scale=1.0,size=5)
ran_2 = np.random.normal(scale=2.0,size=10)
ran_3 = np.random.normal(scale=3.0,size=100)
y_1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(ran_1)**2)
y_2 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(ran_2)**2)
y_3 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(ran_3)**2)
print(y_1.mean(), y_2.mean(), y_3.mean())
print(y_1.std(), y_2.std(), y_3.std())
print(format(kurtosis(y_1)), format(kurtosis(y_2)), format(kurtosis(y_3)))
print(format(skew(y_1)), format(skew(y_2)), format(skew(y_3)))
x = [y_1.mean(), y_1.std(), format(kurtosis(y_1)), format(skew(y_1))]
y = [y_2.mean(), y_2.std(), format(kurtosis(y_2)), format(skew(y_2))]
z = [y_3.mean(), y_3.std(), format(kurtosis(y_3)), format(skew(y_3))]
plt.plot(x,y,z)
plt.show()
```
